```python
import os
import shutil
import sys

# 1. Clear working directory
!rm -rf /kaggle/working/*

# 2. These are the specific absolute paths you provided
absolute_scripts = [
    '/kaggle/input/vesuviusv01/evaluate.py',
    '/kaggle/input/vesuviusv01/prepare_kaggle_data.py',
    '/kaggle/input/vesuviusv01/train.py'
]

WORKING_DIR = '/kaggle/working'

# 3. Copy each script to the working directory
for src_path in absolute_scripts:
    if os.path.exists(src_path):
        # Extract just the filename (e.g., 'train.py') to use as the destination
        filename = os.path.basename(src_path)
        dest_path = os.path.join(WORKING_DIR, filename)
        shutil.copy(src_path, dest_path)
        print(f"‚úÖ Copied {filename}")
    else:
        print(f"‚ùå Script not found at: {src_path}")

# 4. Add working directory to path so scripts can find each other
sys.path.append(WORKING_DIR)
```
